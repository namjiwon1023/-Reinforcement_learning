```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
BC_scores = np.loadtxt('SAC_BC_r.txt')
sac_scores = np.loadtxt('SAC_r.txt')
#TD3_scores = np.loadtxt('TD3_r.txt')
```

```python
def plot_learning_curve(x, y, bc_scores, sac_scores):
    running_avg1 = np.zeros(len(bc_scores))
    for i in range(len(running_avg1)):
        running_avg1[i] = np.mean(bc_scores[max(0, i-10):(i+1)])
        
    running_avg2 = np.zeros(len(sac_scores))
    for i in range(len(running_avg2)):
        running_avg2[i] = np.mean(sac_scores[max(0, i-10):(i+1)])
        
#     running_avg3 = np.zeros(len(TD3_scores))
#     for i in range(len(running_avg3)):
#         running_avg3[i] = np.mean(TD3_scores[max(0, i-10):(i+1)])
        
    plt.plot(x, running_avg1,'r-',label='Behavioral Cloning-SAC')
    plt.plot(y, running_avg2,'b-',label='Soft Actor-Critic')
    # plt.plot(z, running_avg3,'g-',label='Twin Delayed DDPG')
    plt.title('Running average of previous 10 scores')
    plt.legend(loc="best", shadow=True)
```

```python
plt.figure(figsize=(20, 10))
x = [i+1 for i in range(len(BC_scores))]
y = [i+1 for i in range(len(sac_scores))]
# z = [i+1 for i in range(len(TD3_scores))]
plot_learning_curve(x, y, BC_scores,sac_scores)
```
