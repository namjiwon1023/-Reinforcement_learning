```python
import matplotlib.pyplot as plt
import os
import numpy as np
import seaborn as sns

DQN_avg =  np.loadtxt("DQN_avg_reward.txt", delimiter=',')
DQN_total_reward =  np.loadtxt("DQN_Total_reward.txt", delimiter=',')
DQN = DQN_total_reward.mean()

DDQN_avg =  np.loadtxt("DDQN_avg_reward.txt", delimiter=',')
DDQN_total_reward =  np.loadtxt("DDQN_Total_reward.txt", delimiter=',')
DDQN = DDQN_total_reward.mean()

D3QN_avg =  np.loadtxt("avg_reward_D3QN.txt", delimiter=',')
D3QN_total_reward =  np.loadtxt("Total_reward_D3QN.txt", delimiter=',')
D3QN = D3QN_total_reward.mean()

```

```python
print('DQN_avg : ',DQN_avg)
print('DQN_total_reward : ',DQN_total_reward)
print('DDQN_avg : ',DDQN_avg)
print('DDQN_total_reward : ',DDQN_total_reward)
print('D3QN_avg : ',D3QN_avg)
print('D3QN_total_reward : ',D3QN_total_reward)
```

```python
plt.figure()
plt.title("DQN Average reward")
plt.grid(True)

plt.xlabel("Episode")
plt.xlim(0,150)

plt.ylabel("DQN Average reward")
plt.ylim(0, 4)

plt.plot(DQN_avg, "r-", linewidth=2.0, label="DQN Average reward")

plt.legend(loc="best", shadow=True)
plt.savefig('DQN Average reward.png')

plt.show()
```

```python
plt.figure()
plt.title("DDQN Average reward")
plt.grid(True)

plt.xlabel("Episode")
plt.xlim(0,150)

plt.ylabel("DDQN Average reward")
plt.ylim(0, 4)

plt.plot(DDQN_avg, "b-", linewidth=2.0, label="DDQN Average reward")

plt.legend(loc="best", shadow=True)

plt.savefig('DDQN Average reward.png')


plt.show()
```

```python
plt.figure()
plt.title("D3QN Average reward")
plt.grid(True)

plt.xlabel("Episode")
plt.xlim(0,150)

plt.ylabel("D3QN Average reward")
plt.ylim(0, 4)

plt.plot(D3QN_avg, "g-", linewidth=2.0, label="D3QN Average reward")

plt.legend(loc="best", shadow=True)

plt.savefig('D3QN Average reward.png')


plt.show()
```

```python

x = np.linspace(0, 10000)
y1 = x *0 + DQN
y2 = x *0 + DDQN
y3 = x *0 + D3QN

plt.figure(figsize=(22,10))
plt.title("total_reward")
plt.grid(True)

plt.xlabel("Episode")
plt.xlim(0,150)

plt.ylabel("total_reward")
plt.ylim(0, 500)

plt.plot(D3QN_total_reward, "r-", linewidth=2.0, label="D3QN_total_reward")
plt.plot(DDQN_total_reward, "b-", linewidth=2.0, label="DDQN_total_reward")
plt.plot(DQN_total_reward, "g-", linewidth=2.0, label="DQN_total_reward")

plt.plot(x,y1, "k--", linewidth=2.0, label="DQN")
plt.plot(x,y2, "c--", linewidth=2.0, label="DDQN")
plt.plot(x,y3, "m--", linewidth=2.0, label="D3QN")


plt.legend(loc="best", shadow=True)

plt.savefig('Compare_total_reward.png')


plt.show()
```
