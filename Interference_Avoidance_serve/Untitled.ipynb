```python
import matplotlib.pyplot as plt
import os
import numpy as np
avg =  np.loadtxt("avg_reward.txt", delimiter=',')
```

```python
print(avg)
```

```python
plt.figure()
plt.title("Avg_scores")
plt.grid(True)

plt.xlabel("Episode")
plt.xlim(0,10000)

plt.ylabel("Total reward")
plt.ylim(0, 4)

plt.plot(avg, "b-", linewidth=2.0, label="Avg_scores")

plt.legend(loc="upper left", shadow=True)


plt.show()
```
