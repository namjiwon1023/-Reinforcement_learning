```python
import matplotlib.pyplot as plt
import os
import numpy as np
A2C =  np.loadtxt("A2C.txt", delimiter=',')
DDPG =  np.loadtxt("DDPG_scores.txt", delimiter=',')
```

```python
x = np.linspace(0, 2000)
y = x *0 + 2000

plt.figure()
plt.title("Total_scores")
plt.grid(True)

    # 设置X轴
plt.xlabel("Episode")
plt.xlim(0,1000)
# plt.xticks(np.linspace(-4, 4, 9, endpoint=True))

    # 设置Y轴
plt.ylabel("Total reward")
plt.ylim(0, 10000)
# plt.yticks(np.linspace(-1, 1, 9, endpoint=True))

    # 画两条曲线
plt.plot(A2C, "b-", linewidth=2.0, label="A2C_scores")
plt.plot(x,y, "r--", linewidth=2.0, label="mini_scores")

    # 设置图例位置,loc可以为[upper, lower, left, right, center]
plt.legend(loc="upper left", shadow=True)

    # 图形显示
plt.show()
```

```python
import matplotlib.pyplot as plt
import os
import numpy as np
New_DDPG =  np.loadtxt("test.txt", delimiter=',')
Old_DDPG = np.loadtxt("Old_DDPG_scores.txt", delimiter=',')
```

```python
x = np.linspace(0, 2000)
y = x *0 + 2000

plt.figure(figsize=(22,10))
plt.title("Compare",fontdict={'size':42,'weight':'bold'})
plt.grid(True)

plt.xticks(fontsize=32,fontweight='bold')
plt.yticks(fontsize=32,fontweight='bold')
plt.tick_params(labelsize=32)

plt.xlabel("Episode",fontdict={'size':42,'weight':'bold'})
plt.xlim(0, 550)

plt.ylabel("Total Reward",fontdict={'size':42,'weight':'bold'})
plt.ylim(0, 381500)

plt.plot(New_DDPG, "b-", linewidth=2.0, label="New Reward function")
plt.plot(Old_DDPG, "r-", linewidth=2.0, label="Old Reward function")

plt.legend(loc="upper left", shadow=True, fontsize=36)
plt.savefig('New and Old Reward Compare.png')
plt.show()
```
