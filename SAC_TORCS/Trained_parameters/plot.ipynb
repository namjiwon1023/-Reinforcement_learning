```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
bc_scores_new = np.loadtxt('BC_reward_new.txt')
sac_scores_new = np.loadtxt('SAC_r_new.txt')
ddpg_scores_new = np.loadtxt('New_DDPG_scores.txt')
ddpg_scores_old = np.loadtxt('Old_DDPG_scores.txt')
ddpg_scores_old = ddpg_scores_old[:535]
```

```python
def plot_learning_curve(x, y, z, w, bc_scores, sac_scores, ddpg_new, ddpg_old):
    plt.xlabel("Episode", fontsize=50)
    plt.xlim(0,750)
    plt.ylabel("Running average of previous 10 scores", fontsize=50)
    plt.ylim(0, 100000)
    running_avg1 = np.zeros(len(bc_scores))
    for i in range(len(running_avg1)):
        running_avg1[i] = np.mean(bc_scores[max(0, i-10):(i+1)])
        
    running_avg2 = np.zeros(len(sac_scores))
    for i in range(len(running_avg2)):
        running_avg2[i] = np.mean(sac_scores[max(0, i-10):(i+1)])
        
    running_avg3 = np.zeros(len(ddpg_new))
    for i in range(len(running_avg3)):
         running_avg3[i] = np.mean(ddpg_new[max(0, i-10):(i+1)])
            
    running_avg4 = np.zeros(len(ddpg_old))
    for i in range(len(running_avg4)):
         running_avg4[i] = np.mean(ddpg_old[max(0, i-10):(i+1)])
    
    plt.plot(x, running_avg1,'r-',linewidth=5,label='Behavioral Cloning-SAC')
    plt.plot(y, running_avg2,'g-',linewidth=5,label='Soft Actor-Critic')
    plt.plot(z, running_avg3,'b-',linewidth=5,label='DDPG new R func')
    plt.plot(w, running_avg4,'c-',linewidth=5,label='DDPG old R func')
    
    plt.title('Compare',fontsize=50)
    plt.legend(loc="best", shadow=True,fontsize=30)
    plt.xticks(fontsize=50)
    plt.yticks(fontsize=50)
    plt.savefig('Compare')
```

```python
plt.figure(figsize=(30, 15))
x = [i+1 for i in range(len(bc_scores_new))]
y = [i+1 for i in range(len(sac_scores_new))]
z = [i+1 for i in range(len(ddpg_scores_new))]
w = [i+1 for i in range(len(ddpg_scores_old))]
plot_learning_curve(x, y, z, w, bc_scores_new, sac_scores_new, ddpg_scores_new, ddpg_scores_old)
```
