```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
rewardDict =  np.loadtxt("./Total_reward.txt", delimiter=',')
```

```python
print(rewardDict)
```

```python
R_mean = rewardDict.mean()
```

```python
print(R_mean)
```

```python
x = np.linspace(0, 10000)
y = x *0 + R_mean

plt.figure(figsize=(22,10))
plt.title("Compare",fontdict={'size':42,'weight':'bold'})
plt.grid(True)

    # 设置X轴
plt.xlabel("Episode")
plt.xlim(0, 10000)
# plt.xticks(np.linspace(-4, 4, 9, endpoint=True))

    # 设置Y轴
plt.ylabel("Total reward")
plt.ylim(-5, 400)

# plt.yticks(np.linspace(-1, 1, 9, endpoint=True))

plt.plot(rewardDict, "b-", linewidth=2.0, label="scores")
plt.plot(x,y, "r--", linewidth=2.0, label="avg_scores")
plt.legend(loc="lower right", shadow=True)

plt.show()
```
