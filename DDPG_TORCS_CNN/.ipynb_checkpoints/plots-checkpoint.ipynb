```python
import matplotlib.pyplot as plt
import os
import numpy as np
actor_loss =  np.loadtxt("actor_losses.txt", delimiter=',')
critic_loss =  np.loadtxt("critic_losses.txt", delimiter=',')
total_reward = np.loadtxt("Total_scores.txt", delimiter=',')
```

```python
print(total_reward)
```

```python
plt.figure(figsize=(20,10))
plt.title("Total_scores")
plt.grid(True)

plt.xlabel("Episode")
plt.xlim(0,9400)
# plt.xticks(np.linspace(-4, 4, 9, endpoint=True))

plt.ylabel("Total reward")
plt.ylim(0, 525000)
# plt.yticks(np.linspace(-1, 1, 9, endpoint=True))

plt.plot(total_reward, "r-", linewidth=2.0, label="DDPG_CNN_scores")

plt.legend(loc="upper left", shadow=True)
plt.savefig('CNN_DDPG_total_reward.png')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.title("actor_loss")
plt.grid(True)

plt.xlabel("Episode")
plt.xlim(0,9400)


plt.ylabel("actor_loss")
plt.ylim(-14000, 10)


plt.plot(actor_loss, "b-", linewidth=2.0, label="DDPG_CNN_actor_loss")

plt.legend(loc="upper left", shadow=True)
plt.savefig('CNN_DDPG_actor_loss.png')

plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.title("critic_loss")
plt.grid(True)

plt.xlabel("Episode")
plt.xlim(0,9400)

plt.ylabel("critic_loss")
plt.ylim(0, 1950000)

plt.plot(critic_loss, "b-", linewidth=2.0, label="DDPG_CNN_critic_loss")

plt.legend(loc="upper left", shadow=True)
plt.savefig('CNN_DDPG_critic_loss.png')
plt.show()
```
